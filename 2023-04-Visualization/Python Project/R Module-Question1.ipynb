```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('2018 Data.xlsx')
```

```python
plt.scatter('Unemployment Rate','Obesity Rates', color="yellow", marker="*", data=df)

plt.title('Unemployment Rates Compared to Obesity Rates')
plt.xlabel('Unemployment Rate')
plt.ylabel('Obesity Rates')
plt.grid(visible=None, which='major', axis='both', linestyle='-', linewidth=.25)
plt.ylim(bottom=15,top=45)
plt.xlim(left=0,right=18)

plt.show()
```
