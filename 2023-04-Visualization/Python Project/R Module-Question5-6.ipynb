```python
import pandas as pd
from datetime import datetime
```

```python
df_broward = df_all[df_all['County'].isin(['Broward'])]
df_pb = df_all[df_all['County'].isin(['Palm Beach'])]
```

```python
print(df_broward.head)
print(df_pb.head)
```

```python
df_broward_gby = df_broward.groupby(['EventDate']).count() 
print(df_broward_gby.head)
```

```python
df_pb_gby = df_pb.groupby(['EventDate']).count() 
print(df_pb_gby.head)
```

```python
print(df_broward_gby.head)
print(df_pb_gby.head)
```

```python
from datetime import datetime
df_broward_gby.index = pd.to_datetime(df_broward_gby.index, format="%m/%d/%Y")
df_pb_gby.index = pd.to_datetime(df_pb_gby.index, format="%m/%d/%Y")
```

```python
df_broward_sorted = df_broward_gby.sort_index(axis='index', ascending=True)
df_broward_sorted.head
```

```python
df_pb_sorted = df_pb_gby.sort_index(axis='index', ascending=True)
df_pb_sorted.head
```

```python
df_broward_sorted['Date'] = df_broward_sorted.index.values
df_broward_sorted['Daily Cases'] = df_broward_sorted['County']
df_broward_sorted['Moving Average'] = df_broward_sorted['County'].rolling(5, min_periods=0).mean()
print(df_broward_sorted.head)
```

```python
df_pb_sorted['Date'] = df_pb_sorted.index.values
df_pb_sorted['Daily Cases'] = df_pb_sorted['County']
df_pb_sorted['Moving Average'] = df_pb_sorted['County'].rolling(5, min_periods=0).mean()
print(df_pb_sorted.tail)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2,2,figsize=[10,10])
```

```python
plt.sca(axes[0,0])
plt.plot(df_broward_sorted['Date'], df_broward_sorted['Daily Cases'], color='magenta')
axes[0,0].set_title('Daily Number of Cases - Broward')
axes[0,0].set_xlabel('Date')
axes[0,0].set_xticks([])

plt.sca(axes[0,1])
plt.plot(df_broward_sorted['Date'], df_broward_sorted['Moving Average'], color='purple')
axes[0,1].set_title('Moving Average of Daily Cases - Broward')
axes[0,1].set_xlabel('Date')
axes[0,1].set_xticks([])

plt.sca(axes[1,0])
plt.plot(df_pb_sorted['Date'], df_pb_sorted['Daily Cases'], color='magenta')
axes[1,0].set_title('Daily Number of Cases - Palm Beach')
axes[1,0].set_xlabel('Date')
axes[1,0].set_xticks([])

plt.sca(axes[1,1])
plt.plot(df_pb_sorted['Date'], df_pb_sorted['Moving Average'], color='purple')
axes[1,1].set_title('Moving Average of Daily Cases - Palm Beach')
axes[1,1].set_xlabel('Date')
axes[1,1].set_xticks([])

fig.suptitle('Covid Information for Broward and Palm Beach Counties')
plt.show()
```
