```python
import pandas as pd
```

```python
df = pd.read_excel('Simplified Car Sales Data.xlsx')
```

```python
df
```

```python
# Create a (horizontal) bar chart (not a column chart) that shows the total sales and total costs for 
# the United Kingdom, Germany, and France (10 points).
```

```python
uk = df[df['CountryName'] == 'United Kingdom']
ger= df[df['CountryName'] == 'Germany']
fra= df[df['CountryName'] == 'France']
```

```python
uk_totals = uk.sum(axis=0, numeric_only = True)
ger_totals = ger.sum(axis=0, numeric_only = True)
fra_totals = fra.sum(axis=0, numeric_only = True)
```

```python
print(uk_totals)
print(ger_totals)
print(fra_totals)
```

```python
uk_totals_sales = round(uk_totals.SalePrice/1000,0)
uk_totals_costs = round((uk_totals.CostPrice + uk_totals.LaborCost)/1000,0)

ger_totals_sales = round(ger_totals.SalePrice/1000,0)
ger_totals_costs = round((ger_totals.CostPrice + ger_totals.LaborCost)/1000,0)

fra_totals_sales = round(fra_totals.SalePrice/1000,0)
fra_totals_costs = round((fra_totals.CostPrice + fra_totals.LaborCost)/1000,0)
```

```python
totals = pd.DataFrame([('UK',uk_totals_sales, uk_totals_costs),
                         ('Germany', ger_totals_sales,ger_totals_costs),
                         ('France', fra_totals_sales,fra_totals_costs)],
                        columns = ['Country', 'Total Sales', 'Total Costs'])
```

```python
plot_title='Sales and Costs Totals for France, Germany, and the UK in Thoursands of $'
my_colors = ['purple', 'b']

totals.plot('Country', ['Total Sales', 'Total Costs'], kind='barh', title=plot_title,color=my_colors)
```
